```python
df2 = pd.read_excel("dataset_with_errors.csv", sheet_name = "edittmt")
df2.head()
```

```python
len(df2)
```

```python
df2['UnitKerja'] = df2['UnitKerja'].str.lower()
```

```python
df2.groupby('UnitKerja').size().reset_index(name='Jumlah')
```

```python
df2.groupby('UnitKerja').size().reset_index(name='Jumlah').sort_values('Jumlah', ascending=False)
# Karyawan terbanyak ada pada divisi riset yaitu 108 orang
```

```python
df2['TahunMasuk'] = df2['TMTAwal'].dt.year
```

```python
df2.groupby('TahunMasuk').size().reset_index(name='Jumlah').sort_values('TahunMasuk')
```

```python
df2['TahunKeluar'] = df2['TMTAkhir'].dt.year
```

```python
df2.groupby('TahunKeluar').size().reset_index(name='Jumlah').sort_values('TahunKeluar')
```

```python
df2.groupby('TahunKeluar').size().reset_index(name='Jumlah').sort_values('Jumlah', ascending=False)
# Tahun 2022 adalah tahun dimana paling banyak karyawan yang keluar, yaitu sebanyak 49 orang
# Maka tahun 2022 tersebut dapat dianalisis lebih lanjut lagi untuk mengetahui keadaan karyawan di perusahaan tersebut
```
